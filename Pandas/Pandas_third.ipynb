```python
import pandas as pd
import numpy as np
```

```python
def fetch_financial_data(company="META"):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data("META")
df.head()
```

```python
#Do pliku Exel
df.to_csv("fb.csv")
```

```python
df_now = df[(df.index.month ==11) & (df.index.year==2024)]
df_now.to_csv("fb2024.csv")
```
